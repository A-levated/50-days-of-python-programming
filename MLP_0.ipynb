```python
#Two interfaces for the price of one
plt.figure() # create a plot figure

#create the first of the two panels
plt.subplot(2, 1, 1) #(rows, columns, panel number)
plt.plot(x, np.sin(x))

#create the second of the two panels and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
#Object Oriented Interface
#creating a grid of plots with ax as an array of ofntwo axes objects
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
#Adjusting the plot: Line and Colors
plt.plot(x, np.sin(x - 0), color='blue') #specify color by name
plt.plot(x, np.sin(x - 1), color='g')    #shrt color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') #Greyscale between 0 and 1
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) #RGB tuple
plt.plot(x, np.sin(x - 5), color='chartreuse') #All HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='-')
plt.plot(x, x + 1, linestyle='--')
plt.plot(x, x + 2, linestyle='-.')
plt.plot(x, x + 3, linestyle=':')
plt.plot(x, x + 4, linestyle='solid')
plt.plot(x, x + 5, linestyle='dashed')
plt.plot(x, x + 6, linestyle='dashdot')
plt.plot(x, x + 7, linestyle='dotted')
```

```python
#Adjusting the Plot: Axes Limits
plt.plot(x, np.sin(x))

plt.xlim(-2, 12)
plt.ylim(-2, 2);
```

```python
#Reverse plotting
plt.plot(x, np.sin(x))

plt.xlim(12, -2)
plt.ylim(2, -2);
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```python
#Labeling Plots
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000* rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
           cmap='viridis')
plt.colorbar();
```

```python
#Example with the Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features [0], features[1], alpha=0.2,
           s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
