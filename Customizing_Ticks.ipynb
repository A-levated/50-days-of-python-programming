```python
#Hiding ticks or labels
#Using plt.NullLocator()
ax = plt.axes()
ax.plot(np.random.rand(50))

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_locator(plt.NullLocator())
```

```python
fig, ax = plt.subplots(5, 5, figsize=(5, 5))
fig.subplots_adjust(hspace=0, wspace=0)

#Get some face data from Scikit learn
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images

for i in range(5):
    for j in range (5):
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10 * i + j], cmap='bone');
```

```python
#Reducing and increasing the Number of Ticks
fig, ax = plt.subplots(4, 4, sharex=True, sharey=True)
```

```python
#Apply the MaxNLocator to specify max number of ticks to be displayed.
#For every axis set the x and y major locator
for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(3))
    axi.yaxis.set_major_locator(plt.MaxNLocator(3))

fig
```

```python
#Fancy Ticks Formats 
#plot a sine and cosine curve
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x), lw=3, label='Cosine')

#Set up grid, legend and limits 
ax.grid(True)
ax.legend(frameon=False)
ax.axis('equal')
ax.set_xlim(0, 3 * np.pi);
```

```python
#Apply the MultipleLocator
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
fig
```

```python
#Apply the plt.FuncFormatter
def format_func(value, tick_number):
    #find the number of multiples in pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"

    elif N == 1:
        return r"$\pi/2$"

    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)

    else:
        return r"${0}\pi$".format(N // 2)

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
```

```python
#Customizing Matplotlib: Configurations and Stylesheets
plt.style.use('classic')
x = np.random.randn(1000)
plt.hist(x);
```

```python
#Lets adjust the above histogram

# use a gray background
fig, ax = plt.subplots()
ax.set_facecolor('#E6E6E6')
ax.set_axisbelow(True)

#draw solid white gridlines
plt.grid(color='w', linestyle='solid')

#hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)

# hide top and right ticks
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

# lighten ticks and labels
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')

for tick in ax.get_yticklabels():
    tick.set_color('gray')

#control face and edge color of histogram
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    ax[0].hist(np.random.rand(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
    
```

```python
#reset rcParams
hist_and_lines()
```
